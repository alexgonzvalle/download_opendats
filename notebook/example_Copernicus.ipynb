## Descargar datos de Copernicus

```python
import copernicusmarine as cm
cm.login('agonzalezvalla', 'AleX15121523')
```

**Visualizo los productos disponibles**

```python
catalog = cm.describe()
for i, c in enumerate(catalog.products):
    print(i, c.product_id)
```

**Visualizo los datasets disponibles**

```python
for dataset in catalog.products[42].datasets:
    print(dataset.dataset_id)
```

**Leer datos como un dataset**

```python
ds = cm.open_dataset(
    dataset_id = "cmems_mod_glo_wav_anfc_0.083deg_PT3H-i",
    minimum_longitude=-4.7,
    maximum_longitude=-3.3,
    minimum_latitude=43.3,
    maximum_latitude=45.5,
    start_datetime = "2025-01-01 00:00:00",
    end_datetime = "2025-01-08 00:00:00",
    # variables = ["CHL"],
)
ds
```
